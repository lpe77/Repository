### C6. Your task is to use the pairplot from Seaborn and produce 3 plots to visually investigate the relation between the Agricultural area of communes and their Settlement area, Wooded area and the Unproductive area.

```python
sns.pairplot(df, height=5,
             x_vars=["Settlement area", "Wooded area", "Unproductive area"], 
             y_vars=["Agricultural area"],
             kind="reg", plot_kws={'line_kws':{'color':'red'}})
plt.show()
```

#### Are the relations linear?

```python
import numpy as np

def fn_describe_coef(coef):
    if coef == 0:
        msg = str(coef) + ": No linear relationship"
    elif np.abs(coef) > 0 and np.abs(coef) < 0.3:
        msg = str(coef) + ": Weak positive/negative linear relationship"
    elif np.abs(coef) >= 0.3 and np.abs(coef) < 0.7:
        msg = str(coef) + ": Moderate positive/negative linear relationship"
    elif np.abs(coef) >= 0.7 and np.abs(coef) < 1:
        msg = str(coef) + ": Strong positive/negative linear relationship"
    elif np.abs(coef) == 1:
        msg = str(coef) + "Perfect positive/negative linear relationship"
    return msg
```

```python
coef1 = df["Agricultural area"].corr(df["Settlement area"])
"coef1 -> " + fn_describe_coef(coef1)
```

```python
coef2 = df["Agricultural area"].corr(df["Wooded area"])
"coef2 -> " + fn_describe_coef(coef2)
```

```python
coef3 = df["Agricultural area"].corr(df["Unproductive area"])
"coef3 -> " + fn_describe_coef(coef3)
```

# D. Probabilities

```python
total_2000 = df[df["Elevation"]>2000]

total_cnt = len(total_2000)

valais_cnt = len(total_2000[total_2000["Canton"]=="VS"])

probability = valais_cnt/total_cnt

print("Probability that a randomly selected commune with elevation over 2000 is from the canton of Valais: ", probability)
```

```python
import random

total_2000 = df[df["Elevation"]>2000]
experiments = 10000
counter = 0
for i in range(experiments):
    rnd_canton = total_2000["Canton"].sample().values[0]
    if rnd_canton == "VS":
        counter += 1

probability = counter / experiments
print("Experiment probability that a randomly selected commune with elevation over 2000 is from the canton of Valais: ", probability)
```
